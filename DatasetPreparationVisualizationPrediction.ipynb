### **Step 2) Choosing a classifier and fine tuning it**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
target=df['diagnosis']
features_train, features_test, target_train, target_test = train_test_split(features_all,target, test_size=0.25,shuffle=False,random_state=42)
features1_train, features1_test, target1_train, target1_test = train_test_split(features_best,target, test_size=0.25,shuffle=False,random_state=42)
features2_train, features2_test, target2_train, target2_test = train_test_split(features_cp,target, test_size=0.25,shuffle=False,random_state=42)
```

```python
forest = RandomForestClassifier(criterion='gini',n_estimators=100,random_state=42,max_depth=None)
forest.fit(features_train,target_train)
pred = forest.predict(features_test)
print('Accuracy of standard parameters model with all features: %.3f' % accuracy_score(target_test, pred))

forest.fit(features1_train,target1_train)
pred = forest.predict(features1_test)
print('Accuracy of standard parameters model with best features: %.3f' % accuracy_score(target1_test, pred))

forest.fit(features2_train,target2_train)
pred = forest.predict(features2_test)
print('Accuracy of standard parameters model with concave points features: %.3f' % accuracy_score(target2_test, pred))
#all features most accurate
```

```python
forest.fit(features_train,target_train)
pred = forest.predict(features_test)
print('Accuracy of standard parameters model with all features: %.3f' % accuracy_score(target_test, pred))

critEmodel=RandomForestClassifier(criterion='entropy',random_state=42)
critEmodel.fit(features_train,target_train)
predE = critEmodel.predict(features_test)
print('Accuracy of criterion entropy: %.3f' % accuracy_score(target_test, predE))

Hmodel=RandomForestClassifier(n_estimators=100,random_state=42)
Hmodel.fit(features_train,target_train)
predH = Hmodel.predict(features_test)
print('Accuracy of 100 estimators: %.3f' % accuracy_score(target_test, predH))

Amodel=RandomForestClassifier(n_estimators=1000,random_state=42)
Amodel.fit(features_train,target_train)
predA = Amodel.predict(features_test)
print('Accuracy of 1000 estimators: %.3f' % accuracy_score(target_test, predA))

Tmodel=RandomForestClassifier(n_estimators=1,random_state=42)
Tmodel.fit(features_train,target_train)
predT = Tmodel.predict(features_test)
print('Accuracy of 1 estimator: %.3f' % accuracy_score(target_test, predT))

MDmodel=RandomForestClassifier(max_depth=4,random_state=42)
MDmodel.fit(features_train,target_train)
predMD = MDmodel.predict(features_test)
print('Accuracy of set max depth: %.3f' % accuracy_score(target_test, predMD))
```

```python
#best parameters: estimators best at near default level, too little is inaccurate, too much takes too long to run for same performance
#max depth best to be left at default
#criterion entropy is a little more accurate

BestModel=RandomForestClassifier(n_estimators=100,random_state=42,criterion='entropy')
BestModel.fit(features_train,target_train)
predB = BestModel.predict(features_test)
print('Accuracy of best parameters: %.3f' % accuracy_score(target_test, predB))
```

```python
#max accuracy is 98.6% could be due to anomalies in the data that the model is unable to predict

from sklearn import metrics

print("precision of ",BestModel,':',metrics.precision_score(target_test, predB,pos_label='M'))
print("recall of ",BestModel,':',metrics.recall_score(target_test, predB,pos_label='M'))
print("F1 of ",BestModel,':',metrics.f1_score(target_test, predB,pos_label='M'))
```

### **Step 3) Prediction and Evaluation**





```python
# importing necessary libraries
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# loading the iris dataset
iris = datasets.load_iris()
# X -> features, y -> label
X = iris.data
y = iris.target

# dividing X, y into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

# training a linear SVM classifier
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train)
svm_predictions = svm_model_linear.predict(X_test)

# model accuracy for X_test
accuracy = svm_model_linear.score(X_test, y_test)
print(accuracy)

# creating a confusion matrix
cm = confusion_matrix(y_test, svm_predictions)
ConfusionMatrixDisplay(cm).plot()
```
